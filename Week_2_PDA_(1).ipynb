## Part A, NumPy, arrays and shapes

**Learning goal** understand arrays, dimensions, and shapes.  
**Concepts** ndarray, ndim, shape, simple reductions.


```python

import numpy as np

a1 = np.array([1, 2, 3, 4])
a2 = np.array([[1, 2, 3],
               [4, 5, 6]])

print("a1,", a1, ", ndim,", a1.ndim, ", shape,", a1.shape)
print("a2,\n", a2, "\nndim,", a2.ndim, ", shape,", a2.shape)


print("a1 mean,", a1.mean(), ", a2 sum,", a2.sum())
```

import numpy as np → we are bringing in the NumPy library and giving it the shortcut name np.

a1 = np.array([1, 2, 3, 4]) → creates a 1-dimensional array (like a simple list of numbers).

a2 = np.array([[1, 2, 3], [4, 5, 6]]) → creates a 2-dimensional array (like a small table with 2 rows and 3 columns).

a1.ndim and a2.ndim → tells how many dimensions the array has (1D or 2D).

a1.shape and a2.shape → shows the size of the array (for a2, 2 rows × 3 columns).

a1.mean() → calculates the average value of the numbers in a1.

a2.sum() → calculates the total sum of all numbers in a2.

```python

import numpy as np

arr3d = np.arange(2*3*4).reshape(2, 3, 4)
print("ndim,", arr3d.ndim, ", shape,", arr3d.shape)

mean_axis0 = arr3d.mean(axis=0)
mean_axis2 = arr3d.mean(axis=2)

print("Mean along axis 0, shape", mean_axis0.shape)
print(mean_axis0)

print("Mean along axis 2, shape", mean_axis2.shape)
print(mean_axis2)

```

import numpy as np → load the NumPy library.

np.arange(2*3*4) → creates numbers from 0 up to 23 (total 24 numbers).

.reshape(2, 3, 4) → arranges those 24 numbers into a 3D array with shape (2 blocks, each block has 3 rows and 4 columns).

arr3d.ndim → shows the array has 3 dimensions.

arr3d.shape → shows the exact size: (2, 3, 4).

Mean calculations:

arr3d.mean(axis=0) → takes the average across the first dimension (combining the 2 blocks, result shape is 3×4).

arr3d.mean(axis=2) → takes the average across the last dimension (each row of 4 numbers becomes a single average, result shape is 2×3).

axis tells NumPy which direction to “squash” the numbers when calculating.


## Part B, Pandas, Series and DataFrame

**Learning goal** work with labeled data and quick summaries.  
**Concepts** Series, DataFrame, head, info, describe, add a new column.


```python

import pandas as pd

s = pd.Series([10, 20, 30], index=["Mon", "Tue", "Wed"])
print("Series,\n", s)

data = {
    "Name": ["John", "Ashraf", "Sarah"],
    "Age": [35, 41, 32],
    "City": ["Berlin", "Leeds", "Manchester"]
}
df = pd.DataFrame(data)
print("\nDataFrame head,\n", df.head())
print("\nInfo,")
print(df.info())
print("\nDescribe,\n", df.describe())
```

Series part

pd.Series([10, 20, 30], index=["Mon", "Tue", "Wed"]) → creates a Series (like one column of data with labels).

Values: 10, 20, 30

Index labels: Mon, Tue, Wed

Printing it shows both the values and their custom labels.

DataFrame part

data = {...} → make a dictionary with 3 keys (Name, Age, City). Each key has a list of values.

pd.DataFrame(data) → creates a DataFrame, which is like a table (rows × columns).

df.head() → shows the first 5 rows of the DataFrame (default).

df.info() → gives details about the DataFrame (number of rows, columns, data types, memory usage).

df.describe() → provides statistics for numeric columns (mean, min, max, quartiles, etc.).



Series = single labeled column.

DataFrame = full table of rows and columns.

```python

def age_group(age):
    if age < 35:
        return "Young"
    elif 35 <= age <= 40:
        return "Mid"
    else:
        return "Senior"

df["AgeGroup"] = df["Age"].apply(age_group)
print(df)
print("\nCounts,\n", df["AgeGroup"].value_counts())

```

def age_group(age): ... → defines a function that puts ages into categories:

If age is less than 35 → "Young"

If age is between 35 and 39 → "Mid"

If age is 40 or above → "Senior"

df["Age"].apply(age_group) → goes through each value in the Age column and applies the function to classify it.

df["AgeGroup"] = ... → creates a new column in the DataFrame called AgeGroup with these labels.

df["AgeGroup"].value_counts() → counts how many people fall into each group (Young, Mid, Senior).

This shows how we can add new information (derived column) to a dataset and then summarize categories quickly.


## Part C, SciPy, constants and sparse

**Learning goal** use scientific constants and understand sparse storage.  
**Concepts** SciPy constants, CSR matrix, memory awareness.


```python

import numpy as np
from scipy import constants
from scipy.sparse import csr_matrix

print("pi,", constants.pi, ", one minute in seconds,", constants.minute)

arr = np.array([0, 0, 5, 0, 2, 0, 0, 3])
csr = csr_matrix(arr)
print("CSR nonzero values,", csr.data)
print("Count nonzero,", csr.count_nonzero())
```

from scipy import constants → gives access to ready-made scientific constants like π (pi) or unit conversions.

constants.pi → prints the value of π (3.14159...).

constants.minute → shows how many seconds are in a minute (60).

arr = np.array([0, 0, 5, 0, 2, 0, 0, 3]) → creates an array with mostly zeros and a few non-zero values.

csr_matrix(arr) → converts this into a Compressed Sparse Row (CSR) matrix — a memory-efficient way to store only non-zero values.

csr.data → shows the non-zero values (5, 2, 3).

csr.count_nonzero() → counts how many non-zero values there are (3 in this case).



SciPy constants saves us from memorizing scientific values.

Sparse matrices save memory when most of the data is zero.

```python

arr2 = np.array([0,0,7,0,0,0,2,0,0,0,0,5,0,0,0,0,0,1,0,0])
csr2 = csr_matrix(arr2)

rows, cols = csr2.nonzero()
print("Non zero indices, rows", rows, ", cols", cols)
print("Non zero values,", csr2.data)

```

arr2 = np.array([...]) → creates an array of length 20 with mostly zeros and a few non-zero numbers.

csr2 = csr_matrix(arr2) → converts the array into a Compressed Sparse Row (CSR) matrix, which stores only the positions and values of non-zero elements.

csr2.nonzero() → finds the indices (positions) of all non-zero values in the array.

rows → row indices (since this is 1-D, row = 0).

cols → column indices (the positions of the non-zero values in the array).

csr2.data → shows the actual non-zero values stored.


Instead of storing 20 numbers (most of them zeros), CSR only remembers the positions and the values of the 4 non-zeros (7, 2, 5, 1). This saves memory and speeds up processing in large datasets.


## Part D, SymPy, symbolic expressions

**Learning goal** work with symbols, calculus, and algebra in code.  
**Concepts** symbols, substitution, derivative, integral, solve for critical points.


```python

import sympy as sp

x = sp.symbols("x")
expr = 3*(x**2) + 5
print("Expression,", expr)

print("Value at x=4,", expr.subs(x, 4))

print("Derivative,", sp.diff(expr, x))
print("Integral,", sp.integrate(expr, x))
```

## Part F, Matplotlib, quick plots

**Learning goal** create line, bar, and histogram plots.  
**Concepts** titles, labels, basic chart interpretation.


```python


import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 6)
y = x**2
plt.figure()
plt.plot(x, y, marker="o")
plt.title("y equals x squared")
plt.xlabel("x")
plt.ylabel("y")
plt.show()

cats = ["A", "B", "C"]
vals = [5, 8, 3]
plt.figure()
plt.bar(cats, vals)
plt.title("Category counts")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()


ages = np.random.randint(20, 40, 70)
plt.hist(ages, bins=2, edgecolor='black')
plt.title("Age Distribution of Employees")
plt.xlabel("Age Range")
plt.ylabel("Frequency")
plt.show()
```


## Part G, Seaborn, statistical visualizations

**Learning goal** make quick statistical plots and extract insights.  
**Concepts** histplot with hue, boxplot by category, scatter with trend line.


```python


import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

plt.figure()
sns.histplot(data=tips, x="total_bill", hue="day")
plt.title("Total bill by day")
plt.show()

plt.figure()
sns.boxplot(data=tips, x="day", y="total_bill")
plt.title("Total bill distribution by day")
plt.show()

plt.figure()
sns.regplot(data=tips, x="total_bill", y="tip")
plt.title("Tip versus total bill")
plt.show()
```
