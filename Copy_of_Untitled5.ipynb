```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Name": ["Ali", "Maria", "John", "Sara", "David", "Emma"],
    "Age": [25, 32, 45, 29, 40, 36],
    "Department": ["HR", "Finance", "IT", "HR", "IT", "Finance"],
    "Salary": [3000, 4000, 5000, 3200, 5500, 4200]
}

df = pd.DataFrame(data)

print(df.head(3))
```

```python
print(df.info())
print(df.describe())
```

```python
def age_group(age):
    if age < 30:
        return "Young"
    elif age < 40:
        return "Mid"
    else:
        return "Senior"

df["AgeGroup"] = df["Age"].apply(age_group)

print(df)
```

```python
# Count how many employees are in each AgeGroup.


print(df["AgeGroup"].value_counts())


# Find the average salary per department.

avg_salary = df.groupby("Department")["Salary"].mean()

print(avg_salary)
```

```python
# Bar Chart
avg_salary.plot(kind="bar")
plt.title("Average Salary by Department")
plt.xlabel("Department")
plt.ylabel("Average Salary")
plt.show()
```

```python
# Histogram
plt.hist(df["Age"], bins=5)
plt.title("Distribution of Employee Ages")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# Line Plot
sorted_df = df.sort_values("Age")

plt.plot(sorted_df["Age"], sorted_df["Salary"], marker="o")
plt.title("Salary by Age")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.grid(True)
plt.show()
```
